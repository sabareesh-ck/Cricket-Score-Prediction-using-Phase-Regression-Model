```python

from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer
from sklearn.metrics import r2_score, mean_absolute_error


features = [
    "0-5_over_runs", "5-10_over_runs", "10-15_over_runs", "15-20_over_runs",
    "0-5_over_wicket", "5-10_over_wicket", "10-15_over_wicket", "15-20_over_wicket",
    '0-5_over_dotball', '5-10_over_dotball', '10-15_over_dotball',
       '15-20_over_dotball', '0-5_over_boundaries', '5-10_over_boundaries',
       '10-15_over_boundaries', '15-20_over_boundaries','0-5_over_extra_runs', '5-10_over_extra_runs',
       '10-15_over_extra_runs', '15-20_over_extra_runs'

]
target = "target_runs"

md = md.dropna(subset=[target])


imputer = SimpleImputer(strategy="mean")
X = imputer.fit_transform(md[features])
y = md[target].values


X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)


model = LinearRegression()
model.fit(X_train, y_train)


y_pred = model.predict(X_test)


r2 = r2_score(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
mean_actual = y_test.mean()


accuracy = 100 - ((mae / mean_actual) * 100)


print(f"RÂ² Score: {r2:.4f}")
print(f"Mean Absolute Error (MAE): {mae:.2f}")
print(f"Model Accuracy: {accuracy:.2f}%")
```

```python
over_segments = ["0-5_over_runs", "5-10_over_runs", "10-15_over_runs", "15-20_over_runs"]

results = {}

for segment in over_segments:
    matches = md[md[segment] > 52].copy()
    num_matches = len(matches)

    matches.loc[:, "batting_team"] = matches.apply(
        lambda row: row["team1"] if row["1st_innings"] == 1 else row["team2"], axis=1
    )

    num_wins = matches[matches["winner"] == matches["batting_team"]].shape[0]

    win_percentage = (num_wins / num_matches) * 100 if num_matches > 0 else 0

    results[segment] = {"Matches": num_matches, "Wins": num_wins, "Win Percentage": win_percentage}

print("Segment-wise Performance When Scoring More Than 52 Runs:\n")
for segment, data in results.items():
    print(f"{segment}: Matches = {data['Matches']}, Wins = {data['Wins']}, Win Percentage = {data['Win Percentage']:.2f}%")

```

```python

import numpy as np
import pandas as pd
import math


def custom_round(x):
    fraction = x - math.floor(x)
    if fraction >= 2:
        return math.ceil(x)
    if fraction <= 0.5:
        return math.floor(x)
    else:
        return math.ceil(x)

def get_wickets_per_ball_bowler(de, bowlers, start_over, end_over):
    print(f"\nAnalyzing overs {start_over}-{end_over} for bowlers: {bowlers}")

    filtered_df = de[
        (de['over'] >= start_over) &
        (de['over'] < end_over) &
        (de['bowler'].isin(bowlers)) &
        (de['inning'] < 3)
    ]

    if filtered_df.empty:
        print(f"No data available for selected bowlers in overs {start_over}-{end_over}.")
        return 0

    valid_wickets_df = filtered_df[
        (filtered_df['is_wicket'] == 1) &
        (filtered_df['dismissal_kind'] != 'run out') &
        (filtered_df['extras_type'] != 'no ball')
    ]

    total_wickets = valid_wickets_df.shape[0]

    num_balls_bowled = filtered_df.shape[0]

    wickets_per_ball = total_wickets / num_balls_bowled if num_balls_bowled > 0 else 0

    print(f"Total wickets taken (excluding run-outs, no ball wickets, and super over wickets) in {start_over}-{end_over} overs: {total_wickets}")
    print(f"Total balls bowled in {start_over}-{end_over}: {num_balls_bowled}")
    print(f"Wickets per ball in {start_over}-{end_over}: {wickets_per_ball}")

    return wickets_per_ball

def get_user_input():
    overs = [(0, 5), (5, 10), (10, 15), (15, 20)]
    phase_inputs = {}
    for start, end in overs:
        user_input ='JJ Bumrah'
        bowlers_list = [b.strip() for b in user_input.split(",") if b.strip()]
        phase_inputs[(start, end)] = bowlers_list
    return phase_inputs

user_bowlers = get_user_input()

phase_wickets = []
for (start, end), bowlers in user_bowlers.items():
    wickets_per_ball = get_wickets_per_ball_bowler(de, bowlers, start, end)
    phase_wickets.append(wickets_per_ball)

model_input = np.array(phase_wickets).reshape(1, -1)

print("\nFinal model input values (wickets per ball for each phase):")
for i, (start, end) in enumerate([(0, 5), (5, 10), (10, 15), (15, 20)]):
    print(f"{start}-{end} overs: {phase_wickets[i]}")


print("\nWickets per ball multiplied by 30 and rounded (per phase):")
scaled_phase_wickets = []
for i, (start, end) in enumerate([(0, 5), (5, 10), (10, 15), (15, 20)]):
    multiplied_val = phase_wickets[i] * 30
    rounded_val = custom_round(multiplied_val)
    scaled_phase_wickets.append(rounded_val)
    print(f"{start}-{end} overs: Wickets per ball = {phase_wickets[i]:.6f}, multiplied by 30 = {multiplied_val:.6f}, rounded = {rounded_val}")



```

```python
import numpy as np
import pandas as pd
import math

def custom_round(x):
    fraction = x - math.floor(x)
    if fraction <= 0.5:
        return math.floor(x)
    else:
        return math.ceil(x)

def get_scaled_avg_runs(de, batsmen, start_over, end_over):
    filtered_df = de[
        (de['over'] >= start_over) &
        (de['over'] < end_over) &
        (de['batter'].isin(batsmen)) &
        (de['inning'] < 3)
    ]

    filtered_df = filtered_df[filtered_df['extras_type'] != 'wides']

    if filtered_df.empty:
        return 0

    total_runs = filtered_df['batsman_runs'].sum()
    num_balls_played = filtered_df.shape[0]

    avg_runs_per_ball = total_runs / num_balls_played if num_balls_played > 0 else 0
    return avg_runs_per_ball * 30

def get_extra_runs_per_ball_bowler(de, bowlers, start_over, end_over):
    filtered_df = de[
        (de['over'] >= start_over) &
        (de['over'] < end_over) &
        (de['bowler'].isin(bowlers)) &
        (de['inning'] < 3)
    ]

    if filtered_df.empty:
        return 0

    total_extra_runs = filtered_df['extra_runs'].sum()
    num_balls_bowled = filtered_df.shape[0]
    return total_extra_runs / num_balls_bowled if num_balls_bowled > 0 else 0

def get_dot_balls_per_ball_bowler(de, bowlers, start_over, end_over):
    filtered_df = de[
        (de['over'] >= start_over) &
        (de['over'] < end_over) &
        (de['bowler'].isin(bowlers)) &
        (de['inning'] < 3)
    ]

    if filtered_df.empty:
        return 0

    valid_df = filtered_df[filtered_df['extras_type'] != 'no ball']
    dot_balls_df = valid_df[valid_df['total_runs'] == 0]

    total_dot_balls = dot_balls_df.shape[0]
    num_balls_bowled = valid_df.shape[0]

    return total_dot_balls / num_balls_bowled if num_balls_bowled > 0 else 0


def get_boundaries_per_ball_batsman(de, batsmen, start_over, end_over):
    filtered_df = de[
        (de['over'] >= start_over) &
        (de['over'] < end_over) &
        (de['batter'].isin(batsmen)) &
        (de['inning'] < 3)
    ]

    if filtered_df.empty:
        return 0

    valid_df = filtered_df[filtered_df['extras_type'] != 'no ball']
    boundaries_df = valid_df[valid_df['batsman_runs'].isin([4, 6])]

    total_boundaries = boundaries_df.shape[0]
    num_balls_faced = filtered_df.shape[0]
    return total_boundaries / num_balls_faced if num_balls_faced > 0 else 0

def get_wickets_per_phase(de, bowlers, start_over, end_over):
    filtered_df = de[
        (de['over'] >= start_over) &
        (de['over'] < end_over) &
        (de['bowler'].isin(bowlers)) &
        (de['inning'] < 3)
    ]

    if filtered_df.empty:
        return 0

    valid_wickets_df = filtered_df[
        (filtered_df['is_wicket'] == 1) &
        (filtered_df['dismissal_kind'] != 'run out') &
        (filtered_df['extras_type'] != 'no ball')
    ]

    total_wickets = valid_wickets_df.shape[0]
    num_balls_bowled = filtered_df.shape[0]

    wickets_per_ball = total_wickets / num_balls_bowled if num_balls_bowled > 0 else 0
    wickets_scaled = wickets_per_ball * 30
    return custom_round(wickets_scaled)

def get_user_input():
    return {'batsmen': ['V Kohli','AB de Villiers'], 'bowlers': ['Rashid Khan']}

user_data = get_user_input()

phases = [(0, 5), (5, 10), (10, 15), (15, 20)]
runs_values, extra_values, dot_values, boundary_values, wicket_values = [], [], [], [], []

for start, end in phases:
    runs = get_scaled_avg_runs(de, user_data['batsmen'], start, end)
    extra_runs = custom_round(get_extra_runs_per_ball_bowler(de, user_data['bowlers'], start, end) * 30)
    dot_balls = custom_round(get_dot_balls_per_ball_bowler(de, user_data['bowlers'], start, end) * 30)
    boundaries = custom_round(get_boundaries_per_ball_batsman(de, user_data['batsmen'], start, end) * 30)
    wickets = get_wickets_per_phase(de, user_data['bowlers'], start, end)

    print(f"--- Phase {start}-{end} ---")
    print(f"Runs per 30 balls: {runs}")
    print(f"Extra runs per ball (scaled): {extra_runs}")
    print(f"Dot balls per ball (scaled): {dot_balls}")
    print(f"Boundaries per ball (scaled): {boundaries}")
    print(f"Wickets per phase (scaled): {wickets}\n")

    runs_values.append(runs)
    extra_values.append(extra_runs)
    dot_values.append(dot_balls)
    boundary_values.append(boundaries)
    wicket_values.append(wickets)

model_input = np.array(runs_values + extra_values + dot_values + boundary_values + wicket_values).reshape(1, -1)

print("Final Model Input:")
print(model_input)

predicted_target_runs = np.dot(model_input, final_coefficients) + bias

print("\n--- Final Prediction ---")
print(f"Predicted Target Runs: {predicted_target_runs[0]:.2f}")

```
